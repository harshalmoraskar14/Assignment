```python

def armstrong(num):
    digit = str(num)
    power = len(digit)
    total = sum(int(i)**power for i in digit)
    return num==total
    
n = int(input("Enter a number: "))
if armstrong(n):
    print(f"{n} is an Armstrong number.")
else:
    print(f"{n} is NOT an Armstrong number.")
```

#### 31. Create a class Account with private balance attribute and provide methods to deposit and withdraw 

```python
class Account:
    def __init__(self, balance):
        self.__balance = balance   # private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited {amount}. New balance: {self.__balance}"
        return "Invalid deposit amount"

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew {amount}. New balance: {self.__balance}"
        return "Invalid withdrawal amount"

    def get_balance(self):
        return self.__balance
```

```python
# Example usage
acc = Account(1000)
print(acc.deposit(5000))   # Deposited 500. New balance: 1500
print(acc.withdraw(200))  # Withdrew 200. New balance: 1300
print(acc.get_balance())  # 1300
```

```python


class Bank:
    def __init__(self,account_number, name, balance, pin):
        self.__account_number = account_number
        self.__name = name 
        self.__balance = balance
        self.__pin=pin


    def deposite(self,amount):
        if amount>0:
            self.__balance+=amount
            return f"Deposited {amount}. New balance: {self.__balance}"
        return "Invalid deposit amount"


    def Withdrawal(self,amount,pin):
        if pin!=self.__pin:
            return "Incoorect Pin"
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew {amount}. New balance: {self.__balance}"
        return "Invalid withdrawal amount"



    def show_detail(self,pin):
        if pin==self.__pin:
            return f"Account Number: {self.__account_number}  Name: {self.__name}    Balance: {self.__balance}"
        else:
            return "Access denied! Incorrect PIN."
```

#### 38. Write a program to find the difference in days between two given dates. 

```python

def differnece_date(date1,date2):
    date1 = datetime.strptime(date1,"%d-%m-%Y")
    date2 = datetime.strptime(date2,"%d-%m-%Y")
    diff = abs((date1-date2).days)
    return diff

date1 = input("Enter first date (DD-MM-YYYY): ")
date2 = input("Enter second date (DD-MM-YYYY): ")

print("Difference in days:", differnece_date(date1, date2))
```
