```python
# 2 que

def count_vowels(word):
    """This Function count the vowel in the word."""
    word = word.lower()
    a = e = i = o = u = 0
    for ch in word:
        if ch == "a":
            a += 1
        elif ch == "e":
            e += 1
        elif ch == "i":
            i += 1
        elif ch == "o":
            o += 1
        elif ch == "u":
            u += 1
            
    return {"a": a, "e": e, "i": i, "o": o, "u": u}
```

```python
count_vowels("Harshal")
```

```python
# 38 ( que)

f = open("123.txt", "r")
text = f.read().lower()
f.close()

vowels = 0
consonants = 0

for ch in text:
    if ch.isalpha():
        if ch in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
```

```python
with open("123.txt") as f:
    text = f.read().lower()
vowels = sum(ch in "aeiou" for ch in text)
consonants = sum(ch. isalpha() and ch not in "aeiou" for ch in text)
print("Vowels:",vowels, "Consonants:",consonants)
```

```python
# 49 (que)

try:
    num = int(input("Enter a number: "))   
    result = num + "5"                     # this code will show the error 
    print(result)
except ValueError:
    print("ValueError: Please enter numbers only!")
except TypeError:
    print("TypeError: Cannot add number and string together!")
```
